```python
#4. Visualize from which year the country that has won the most Nobel Prizes started to dominate and 
# what factors contributed to this dominance.
# Could you share your insights?
```

```python
award_counts = df["birth_country"].value_counts()
print(award_counts)
```

```python
most_awards_country = award_counts.idxmax() #idxmax() returns the index of the country that has won the most awards.

print("The country that has won the most awards: ", most_awards_country)
```

```python
#To be able to visualize the data, we convert the ‘year’ column from object to datetime type.
```

```python
#Converting the ‘year’ column from int to datetime.
df["year"] = pd.to_datetime(df["year"], format="%Y")
```

```python
#Let’s convert the ‘birth_date’ column to a datetime object.
df['birth_date'] = pd.to_datetime(df['birth_date']) 
```

```python
df.info()
```

```python
#Filters the data of the country with the most awards
most_awards_df = df[df['birth_country'] == most_awards_country]

print(most_awards_df)
```

```python
# Calculating the number of awards by year.

award_counts_by_year = most_awards_df["year"].value_counts().sort_index()

print(award_counts_by_year)


#First, the expression most_awards_df[“year”] filters the “year” column of the DataFrame named “most_awards_df”.

#Then, the value_counts() function calculates how many times each year is repeated and returns the results as a new dataframe.
#Thus, it creates a dataframe containing the number of awards for each year.

#The sort_index() function sorts the years by their indices.
```

```python
plt.title('The Country with the Most Awards: ' + most_awards_country)
plt.xlabel('Year')
plt.ylabel('Number of Prize')
plt.plot(award_counts_by_year.index, award_counts_by_year.values)
plt.show()
```

In the visual, we see the Nobel Prizes awarded annually to the country with the most awards, under the “birth_country” column. The graph shows the number of awards from 1901 to 2020.

We can observe that the United States of America is the country with the most awards in the analysis. It is noticeable that the United States began to dominate the Nobel Prizes starting from the 1970s. This dominance can be attributed to several factors. One of them could be the significant emphasis placed on technological advancements and scientific research during the Cold War period in the U.S. Another reason might be the presence of researchers in the U.S. who made outstanding contributions across various scientific fields and the robust network of universities. Furthermore, the U.S. offers a competitive scientific environment and a supportive research infrastructure, which are also crucial factors.

```python
#6. I will visualize and interpret the categories and countries of Nobel Prize winners from 1938 to 1945.
                 
```

```python
# Uniqueness of 'year'
df['year'] = pd.to_datetime(df['year'])
print(df["year"].unique())

# Filtering the DataFrame
date_range = df[(df["year"] >= pd.to_datetime("1938-01-01")) & (df["year"] <= pd.to_datetime("1945-12-31"))]

# For the category visualization
plt.figure(figsize=(10, 6))
sns.countplot(data=date_range, x="category")
plt.title("Nobel Prize Categories from 1938 to 1945")
plt.xlabel("Categories")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# For the country visualization
plt.figure(figsize=(20, 10))
sns.countplot(data=date_range, x="birth_country")
plt.title("Nobel Prize Countries from 1938 to 1945")
plt.xlabel("Countries")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Assuming 'df' is already defined and includes the 'year', 'category', and 'birth_country' columns

# Ensuring the 'year' column is in datetime format
df['year'] = pd.to_datetime(df['year'])

# Print unique years to check available data
print(df["year"].unique())

# Filtering the DataFrame for the specified date range
date_range_2 = df[(df["year"] >= pd.to_datetime("1947-01-01")) & (df["year"] <= pd.to_datetime("1991-12-31"))]

# For the category visualization
plt.figure(figsize=(10, 6))
sns.countplot(data=date_range_2, x="category")
plt.title("Categories of Nobel Prize Winners from 1947 to 1991")
plt.xlabel("Categories")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# For the country visualization
plt.figure(figsize=(20, 6))
sns.countplot(data=date_range_2, x="birth_country")
plt.title("Countries of Nobel Prize Winners from 1947 to 1991")
plt.xlabel("Countries")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
#8.I will visualize and comment on the countries and ages of individuals who won the Nobel Prizes in Chemistry, Literature, Peace, Physics and
#Medicine after 2000.
```

```python
# Assuming 'df' is already defined and includes the 'year', 'category', 'birth_country', and 'age' columns

# Ensuring the 'year' column is in datetime format
df['year'] = pd.to_datetime(df['year'])

# Print unique years to check available data
print(df["year"].unique())

# Filtering the DataFrame for the specified date range and categories
filtered_data = df[(df["category"].isin(["Chemistry", "Literature", "Peace", "Physics", "Medicine"])) & 
                   (df["year"] >= pd.to_datetime("2000-01-01"))]

# For the countries visualization
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_data, x="birth_country")
plt.title("The countries of the Nobel Prize winners in Chemistry, Literature, Peace, Physics, and Medicine after 2000.")
plt.xlabel("Countries")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()

# For the ages visualization
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_data, x="age")
plt.title("The ages of the Nobel Prize winners in Chemistry, Literature, Peace, Physics, and Medicine after 2000.")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
